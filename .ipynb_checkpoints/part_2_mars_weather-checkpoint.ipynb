```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
import time
```

```python
# Define the Edge WebDriver path (ensure this is the correct location)
edge_driver_path = r"C:\Users\dalto\Desktop\Application Downloads\Edge\edgedriver_win32\msedgedriver.exe"

# Set up Edge options
options = Options()
options.add_argument("--start-maximized")  # Open browser maximized

# Set up Edge WebDriver service
service = Service(edge_driver_path)

# Initialize WebDriver for Edge
browser = webdriver.Edge(service=service, options=options)
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.get(url)

# Allow time for the page to load
time.sleep(2)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.page_source  # Get the page source
soup = BeautifulSoup(html, "html.parser")

# Extract the table
table = soup.find("table", class_="table")
```

```python
# Extract all rows of data
rows = table.find_all("tr")[1:]  # Skip header row
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
data = []

# Loop through the scraped data to create a list of rows
for row in rows:
    cols = row.find_all("td")
    row_data = [col.text.strip() for col in cols]
    data.append(row_data)

# Define column names
columns = ["id", "terrestrial_date", "sol", "ls", "month", "min_temp", "pressure"]
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
df = pd.DataFrame(data, columns=columns)
```

```python
# Confirm DataFrame was created successfully
df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
df.dtypes
```

```python
# Change data types for data analysis
df["terrestrial_date"] = pd.to_datetime(df["terrestrial_date"])
df["sol"] = df["sol"].astype(int)
df["ls"] = df["ls"].astype(int)
df["month"] = df["month"].astype(int)
df["min_temp"] = df["min_temp"].astype(float)
df["pressure"] = df["pressure"].astype(float)
```

```python
# Confirm type changes were successful by examining data types again
df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
mars_months = df["month"].nunique()
print(f"There are {mars_months} months on Mars.")
```

```python
# 2. How many sols (Martian days) worth of data are there?
martian_days = df["sol"].nunique()
print(f"There are {martian_days} Martian days in the dataset.")
```

```python
# 3. What is the average minimum temperature by month?
temp_by_month = df.groupby("month")["min_temp"].mean()
print(temp_by_month)
```

```python
# Plot the average minimum temperature by month
plt.figure(figsize=(8,5))
temp_by_month.plot(kind="bar")
plt.xlabel("Month")
plt.ylabel("Average Minimum Temperature (°C)")
plt.title("Average Minimum Temperature by Month")
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
sorted_temp_by_month_asc = temp_by_month.sort_values()

# Plot the sorted values
plt.figure(figsize=(8,5))
sorted_temp_by_month_asc.plot(kind="bar")
plt.xlabel("Month")
plt.ylabel("Average Minimum Temperature (°C)")
plt.title("Average Minimum Temperature by Month (Coldest to Hottest)")
plt.show()
```

```python
# 4. What is the average pressure by month?
pressure_by_month = df.groupby("month")["pressure"].mean()
print(pressure_by_month)
```

```python
# Plot the average pressure by month
plt.figure(figsize=(8,5))
pressure_by_month.plot(kind="bar")
plt.xlabel("Month")
plt.ylabel("Average Atmospheric Pressure (Pa)")
plt.title("Average Atmospheric Pressure by Month")
plt.show()
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
sorted_pressure_by_month_asc = pressure_by_month.sort_values()

plt.figure(figsize=(8,5))
sorted_pressure_by_month_asc.plot(kind="bar")
plt.xlabel("Month")
plt.ylabel("Average Atmospheric Pressure (Pa)")
plt.title("Average Atmospheric Pressure by Month")
plt.show()
```

```python
# 5. How many terrestrial (Earth) days are there in a Martian year?
# Visually estimate the length of a Martian year by plotting the daily minimum temperature
plt.figure(figsize=(10,5))
plt.plot(df["terrestrial_date"], df["min_temp"], marker=".", linestyle="-")
plt.xlabel("Martian Date")
plt.ylabel("Minimum Temperature (°C)")
plt.title("Daily Minimum Temperature on Over Time")
plt.xticks(rotation=45)
plt.show()
```

#### Minimum Temperature

> The coldest month on Mars at Curiosity’s location is *Month 3* (Martian March) with an average minimum temperature of around *-83°C*.

> The warmest month is *Month 8* (Martian August), where the average minimum temperature rises to about *-68°C*.

> The temperature variation between the coldest and hottest months is approximately *15°C*, indicating significant seasonal changes.

> The trend suggests that winter occurs around *Month 3* and summer around *Month 8* in Curiosity's location.

- Conclusion: The coldest months align with Martian winter, while the warmest months correspond to Martian summer, likely influenced by Mars' elliptical orbit and axial tilt.

#### Atmospheric Pressure

> The lowest atmospheric pressure occurs in *Month 6* (Martian June), where the average pressure drops to around *750* Pascals.

> The highest atmospheric pressure is observed in *Month 9* (Martian September), where it peaks at around *920* Pascals.

> The fluctuation in pressure is roughly *170* Pascals, showing noticeable seasonal changes.

> The pattern of pressure changes suggests that pressure is lowest during the hottest months and highest during the coldest months.

- Conclusion: The atmospheric pressure on Mars increases in colder months and decreases in warmer months, likely due to the condensation and sublimation of CO₂ at the polar ice caps, which affects global air pressure.

#### Year Length

> The temperature plot over time reveals a repeating cyclical pattern, indicating seasonal changes.

> The cycle from one temperature low (coldest month) to the next suggests that a Martian year lasts approximately *687* Earth days.

> This aligns with scientific data, which states that one Martian year equals about *1.88* Earth years.

- Conclusion: The data confirms that a Martian year is about 687 Earth days, showing that Mars experiences longer seasons compared to Earth due to its greater distance from the Sun and slower orbit.
